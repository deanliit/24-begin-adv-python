```python
#sqlite-demo-1.py

## connect to sqlit database
## ie: mysql, oracle, sql server, postgrsql, other

#run query to display data

import sqlite3

db = "demodb1.db"
sql1 = "select id, name, email, phone from contact"

#connect to other databases: mysql, oracle, sql server postgresql, other
#some dabases has there own libraries ie: import mysql etc...
conn = sqlite3.connect(db)

#following code is same for all dabasses
cur = conn.cursor()
rows=cur.execute(sql1)

rows = cur.fetchall()
#cur.fetchone()
#cur.fetchmany()

conn.close()

for row in rows:
  print(row)
  #print(row[1])
```

```python
#sqlite-demo-2.py

## connect to sqlit database

#load data into pandas dataframe
#run query to display data

import pandas as pd
import sqlite3

db = "demodb1.db"
sql1 = "select id, name, email, phone from contact"

conn = sqlite3.connect(db)
df = pd.read_sql(sql1, conn)
conn.commit


df.head()


```
