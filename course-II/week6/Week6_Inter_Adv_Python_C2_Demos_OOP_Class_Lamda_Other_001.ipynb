## Object Oriented Programming Summary
- re-use code in other projects
- modify and extend code without changing previsous code
- as project requirements increase, add oop to re-use code
- OOP guidlines for OOP code re-use

- How/When OOP is used
-----------------------------
- as project requirements increase, add oop to re-use code
- instead of writing may functions as your application increase
- group functions and data into an object called OOP
- without OOP, code would have many functions and without any grouping
- when create an object in code - object and data in in its own memory
  called instance

OOP guidlines / code re-use
-----------------------------
1. define class/object/template
2. constructor - provide data when object is created in memory
3. this/self - instance - keyword - refence to data when object is in memory
4. inheritance - re-use code from a previous class/object/template
5. polymophism - use same function of inherited code ie: inherited class.func1,
   child/class or sub/class.func1

## Introduction to functional Programming Summary

- func_group_a(data1,...).func_group_b(data...)
- create a function withiin a function
- when function goies out of scope the memory is returned

- How/When Punctional Programming is used
- advanced feature: function can access data above scope or outside the functions ie: closure
- can assign function to a variable
- advanced feature: can save when data inside functions ie: closure

- simplify/reduce code to 1 function instead of code block
- used often in mathematical programming or AI/ML

```python
#week6-oop-demo-3.py

#define class/object/template - deine re-use code ie: data and functions
class Person:
  pass

class Item:
  None

class Car:
 pass

class MyClass:
  x = 5 #can add data and functions

#use class template / instance  - create object in own memory
var1 = Person
var2 = Person
var3 = Car
var4 = Item
var5 = MyClass  #same memory space as myclasss ie: shared memory
var6 = MyClass  #same memory space as myclasss ie: shared memory

var6.x = 10

print("myclass:var5 ", var5.x)
print("myclass:var6 ", var6.x)
```

```python
#week6-oop-demo-4.py

#define class/object/template - deine re-use code ie: data and functions
class Person:
  pass

class Item:
  None

class Car:
 pass

class MyClass:
  x = 5 #can add data and functions

#use class template / instance  - create object in own memory
var1 = Person
var2 = Person
var3 = Car
var4 = Item
var5 = MyClass()  #different or own memory space as myclasss ie: own memory ie: instance/constructor
var6 = MyClass()  #different or own memory space as myclasss ie: own memory ie: instance/constructor

var6.x = 10

print("myclass:var5 ", var5.x)
print("myclass:var6 ", var6.x)
```
